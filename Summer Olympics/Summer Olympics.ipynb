```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#reading csv file
df = pd.read_csv("summer.csv")

#copying all records consisting Medal field as Gold
lst = df[df['Medal'] == 'Gold']
```

## 1. In how many cities Summer Olympics is held so far?

```python
print("\nName of the Cities where Summer Olympics were held:\n",df['City'].unique(),"\n\nTotal number of cities are: ",len(df['City'].unique()))
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
#Since lst consist of Gold medal records
#We will count the number of times a sport has been repeated in the file

data = []
for sport in lst['Sport'].unique() :
    data.append([sport , len(lst[lst['Sport']  == sport])])
data= pd.DataFrame(data,columns = ['Sport','Number of Gold Medals']).sort_values(by='Number of Gold Medals', ascending=False).head().plot(x = 'Sport', y = 'Number of Gold Medals', kind = 'bar', figsize = (15,5))
```

## 3.Which sport is having most number of medals so far? (Top5)

```python
#counting the number of times each sport occurs in list df

data = []
for sport in df['Sport'].unique() :
    data.append([sport , len(df[df['Sport']  == sport])])
data = pd.DataFrame(data,columns = ['Sport','Most Number of Medals']).sort_values(by='Most Number of Medals', ascending=False).head().plot(x = 'Sport', y = 'Most Number of Medals', kind = 'bar', figsize = (15,5))
```

## 4. Which player has won most number of medals?

```python
#counting the of times an athlete name appears in the list df
#since it contains list of players who won medals in summer olympics

data = []
    
for athlete in df['Athlete'].unique() :
    data.append([athlete , len(df[df['Athlete']  == athlete])])
data = pd.DataFrame(data,columns = ['Athlete','Medals Won']).sort_values(by='Medals Won', ascending=False).head().plot(x = 'Athlete', y = 'Medals Won', kind = 'bar', figsize = (15,5))
```

## 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
#Since lst consist records of Gold Medalist
#counting the of times an athlete name appears in the list lst
#since it contains list of players who won gold medals in summer olympics

data = []
for athlete in lst['Athlete'].unique() :
    data.append([athlete , len(lst[lst['Athlete']  == athlete])])
data = pd.DataFrame(data,columns = ['Athlete','Number of Gold Medals']).sort_values(by='Number of Gold Medals', ascending=False).head().plot(x = 'Athlete', y = 'Number of Gold Medals', kind = 'bar', figsize = (5,5))
```

## 6. In which year India won first Gold Medal in Summer Olympics?

```python
#Since lst consist records of Gold Medalist
#We are copying all the records where country is India in the list lsts

lsts = lst[lst['Country'] == 'IND']
print("India's First Gold Medal was in the year : ",lsts['Year'].min())
```

## 7. Which event is most popular in terms on number of players? (Top5)

```python
#Since df contains records of the players who have played an event and won a medal
#So we just need to count the number of times an event occured in the list

data = []
for event in df['Event'].unique() :
    data.append([event , len(df[df['Event']  == event])])
data=pd.DataFrame(data,columns = ['Event','Number of players played']).sort_values(by='Number of players played', ascending=False).head().plot(x = 'Event', y = 'Number of players played', kind = 'bar', figsize = (5,5))
```

## 8.Which sport is having most female Gold Medalists?

```python
#Since lst consist records of Gold Medalist
#We are copying all the records where Gender is Women in the list ls
#So we just need to count the number of times a sport occurs in the list ls

ls = lst[lst['Gender'] == 'Women']
data = []
for women in ls['Sport'].unique() :
    data.append([women , len(ls[ls['Sport']  == women])])
data=pd.DataFrame(data,columns = ['Sport','Number of Females']).sort_values(by='Number of Females', ascending=False).head().plot(x = 'Sport', y = 'Number of Females', kind = 'bar', figsize = (10,5))
```
